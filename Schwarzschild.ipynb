```python
from gr_class import Manifold
import sympy as sp 
```

```python
# Schwarzschild metric

t = sp.Symbol('t')
r = sp.Symbol('r')
theta = sp.Symbol("theta") 
phi = sp.Symbol('phi') 
x = [t,r,theta,phi]

G = sp.Symbol('G')
M = sp.Symbol('M')

g = sp.zeros(4, 4)
g[0, 0] = - (1 - 2 * G * M / r)
g[1, 1] = 1 / (1 - 2 * G * M / r)
g[2, 2] = r ** 2
g[3, 3] = r ** 2 * sp.sin(theta)**2

Schwarzschild = Manifold(x, g)
```

```python
metric = Schwarzschild.get_metric()
inverse_metric = Schwarzschild.get_inverse_metric()
christoffel =  Schwarzschild.get_christoffel()
riemann = Schwarzschild.get_riemann()
kretschmann = Schwarzschild.get_kretschmann()
ricci = Schwarzschild.get_ricci()
scalar = Schwarzschild.get_scalar()
einstein = Schwarzschild.get_einstein()
```
