```python
from gr_class import Manifold
import sympy as sp 
```

```python
# Friedmann-Robertson-Walker metric

t = sp.Symbol('t')
r = sp.Symbol('r')
theta = sp.Symbol("theta") 
phi = sp.Symbol('phi') 
x = [t,r,theta,phi]

a = sp.Function('a')(t)
k = sp.Symbol('k')

g = sp.zeros(4, 4)
g[0, 0] = - 1
g[1, 1] = a**2 / (1 - k * r**2)
g[2, 2] = a**2 * r**2
g[3, 3] = a**2 * r**2 * sp.sin(theta)**2

FRW = Manifold(x, g)
```

```python
metric = FRW.get_metric()
inverse_metric = FRW.get_inverse_metric()
christoffel =  FRW.get_christoffel()
riemann = FRW.get_riemann()
kretschmann = FRW.get_kretschmann()
ricci = FRW.get_ricci()
scalar = FRW.get_scalar()
einstein = FRW.get_einstein()
```
