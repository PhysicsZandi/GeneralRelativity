```python
from gr_class import Manifold
import sympy as sp 
```

```python
# Reissner-Nordstrom metric

t = sp.Symbol('t')
r = sp.Symbol('r')
theta = sp.Symbol("theta") 
phi = sp.Symbol('phi') 
x = [t,r,theta,phi]

Q = sp.Symbol('Q')
m = sp.Symbol('m')

g = sp.zeros(4, 4)
g[0, 0] = - ( 1 - 2 * m / r + Q**2 / r**2 )
g[1, 1] = ( 1 - 2 * m / r + Q**2 / r**2 )**(-1)
g[2, 2] = r**2
g[3, 3] = r**2 * sp.sin(theta)**2

RN = Manifold(x, g)
```

```python
metric = RN.get_metric()
inverse_metric = RN.get_inverse_metric()
christoffel =  RN.get_christoffel()
riemann = RN.get_riemann()
kretschmann = RN.get_kretschmann()
ricci = RN.get_ricci()
scalar = RN.get_scalar()
einstein = RN.get_einstein()
```
